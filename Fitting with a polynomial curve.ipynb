```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [2, 8, 14, 28, 39, 62]

plt.title("x / y")
plt.xlabel("x", loc = "right")
plt.ylabel("y", loc = "top")
plt.grid()
plt.scatter(x, y, color = 'k', alpha = 0.5)
plt.plot(x, y)
plt.legend(['scatter', 'line'])
plt.show()

#Algorithm for finding a straight line using linear regression method

def straight_line_equation(x, y):
    if len(x) == len(y):
        n = len(x)

    sum_x = sum_xy = sum_x2 = sum_y = 0

    for i in range(n):
        sum_x += x[i]
        sum_y += y[i]
        sum_xy += x[i] * y[i]
        sum_x2 += x[i] ** 2

    x_mean = sum_x / n
    y_mean = sum_y / n

    a = ((y_mean * sum_x2) - (x_mean * sum_xy)) / (sum_x2 - (n * (x_mean ** 2)))
    b = (sum_xy - (x_mean * sum_y)) / (sum_x2 - (n * (x_mean ** 2)))
                                  
    print("The straight line equation: y = (%.3f) + (%.3f)x\n" % (a, b))

straight_line_equation(x, y)
                                
straight_line = []

for i in range(n):
    straight_line.append((-3.571) + (11.629) * x[i])
    
#End of Straight-line fitting algorithm

print("Linear fit")
plt.title("x / y")
plt.xlabel("x", loc = "right")
plt.ylabel("y", loc = "top")
plt.grid()
plt.scatter(x, y, color = 'k', alpha = 0.5)
plt.plot(x, straight_line, color = "green")
plt.plot(x, y)
plt.legend(['scatter', 'linear fit', 'allfit'])
plt.show()

```

```python
"""
Fitting with a polynomial curve

The polynomial has the form,

f(x) = a0 + a1x + a2(x ** 2) + ... +  an(x ** n)

If a set of data containing m points is to be fitted by the polynomial curve of degree n, a system of linear
equations is formulated to compute the values of the coefficeints.


Denoted as: [A]{a} = {B} or [A](x) = (b)  "[]" - represents a matrix, "{}" or "()" represents a vector

Consider a matrix-A and a vector-b of nth degree polynomial:

A = [[m             sum(xi)             sum(xi ** 2)       ...  sum(xi ** n)]
     [sum(xi)       sum(xi ** 2)        sum(xi ** 3)       ...  sum(xi ** (n + 1))]
        .
        .
        .
     [sum(xi ** n)  sum(xi ** (n + 1))  sum(xi ** (n + 2)) ... sum(xi ** 2n)]]

b = [[       sum(yi)     ]
     [    sum(xi * yi)   ]
     [sum((xi ** n) * yi)]]
     
     
Consider a matrix-A and a vector-b of 3rd degree polynomial:

A = [[m             sum(xi)      sum(xi ** 2)  sum(xi ** 3)]
    [sum(xi)       sum(xi ** 2)  sum(xi ** 3)  sum(xi ** 4)]
    [sum(xi ** 2)  sum(xi ** 3)  sum(xi ** 4)  sum(xi ** 5)]
    [sum(xi ** 3)  sum(xi ** 4)  sum(xi ** 5)  sum(xi ** 6)]]

b = [[      sum(yi)      ]
     [    sum(xi * yi)   ]
     [sum((xi ** 2) * yi)]
     [sum((xi ** 3) * yi)]]
"""

import numpy as np
x = np.arange(6)
y = np.array([2, 8, 14, 28, 39, 62], float)

#Finding the equation of a polynomial curve for fitting

def polynomial_equation(x, y, nth_order):
    if len(x) == len(y):
        m = len(x)         #number of data-points
    
    n = nth_order                 #degree of polynomial quadratic - 2, cubic - 3, and so on.

    A = np.zeros((n + 1, n + 1))      #size of the matrix A is always (n + 1) x (n + 1)
    B = np.zeros(n + 1)               #the order of B and a are always equal.
    a = np.zeros(n + 1)               #this array contains all the coefficients [a0, a1, a2, ..., an]

    for row in range(n + 1):
        for col in range(n + 1):
            if (row == 0) and (col == 0):
                A [row, col] = m
                continue
            A[row, col] = np.sum(x ** (row + col))
        B[row] = np.sum((x ** row) * y)

    a = np.linalg.solve(A, B)

    print(f"The polynomial: ")
    print("f(x): \t %f" % a[0], end="")
    for i in range(1, n + 1):
        print(' %+ fx^%d' % (a[i], i), end="")
    print("\n")

#End of finding the equation of a ploynomial curve for fitting

#Algorithm for finding a straight line using linear regression method

def straight_line_equation(x, y):
    if len(x) == len(y):
        n = len(x)

    sum_x = sum_xy = sum_x2 = sum_y = 0

    for i in range(n):
        sum_x += x[i]
        sum_y += y[i]
        sum_xy += x[i] * y[i]
        sum_x2 += x[i] ** 2

    x_mean = sum_x / n
    y_mean = sum_y / n

    a = ((y_mean * sum_x2) - (x_mean * sum_xy)) / (sum_x2 - (n * (x_mean ** 2)))
    b = (sum_xy - (x_mean * sum_y)) / (sum_x2 - (n * (x_mean ** 2)))

    print("The straight line equation: y = (%.3f) + (%.3f)x\n" % (a, b))

#End of Straight-line fitting algorithm

polynomial_equation(x, y, 2)
polynomial_equation(x, y, 3)
straight_line_equation(x, y)

straight_line = []
quadratic_curve = []
cubic_curve = []

for i in range(len(x)):
    straight_line.append((-3.571) + (11.629) * x[i])
    quadratic_curve.append(2.678571 + 2.253571 * x[i] + 1.875000 * (x[i] ** 2))
    cubic_curve.append(1.928571 + 5.678571 * x[i] - 0.000000 * (x[i] ** 2) + 0.250000 * (x[i] ** 3))
    
plt.title("x / y")
plt.xlabel("x", loc = "right")
plt.ylabel("y", loc = "top")
plt.grid()
plt.scatter(x, y, color = 'k', alpha = 0.5)
plt.plot(x, straight_line, color = "green")
plt.plot(x, quadratic_curve, color = "orange")
plt.plot(x, cubic_curve, color = "purple")
plt.plot(x, y)
plt.legend(['scatter', 'linear fit', 'quadratic fit', 'cubic fit', 'allfit'])
plt.show()
```
