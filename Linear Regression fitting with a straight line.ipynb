```python
"""
y = a + bx

a = (y_mean * (sum(x ** 2)) - (x_mean * sum(x * y))) / (sum(x ** 2) - (n * (x_mean ** 2))

b = (sum(x * y) - (x_mean * sum(y))) / (sum(x ** 2) - (n * (x_mean ** 2))

x_mean = sum(x) / n

y_mean = sum(y) / n

Here x and y are arrays not constants, n is the number of elements in a list. We would assume that the number of 
elements in a list of both x and y are the same.

This method uses to get a straight line through a linear regression method, through finding the coefficients of a
and b.

"""
import matplotlib.pyplot as plt

x = [-1, 0, 1]
y = [0.5, 1.2, 1.5]

def straight_line_equation(x, y):
    if len(x) == len(y):
        n = len(x)

    sum_x = sum_xy = sum_x2 = sum_y = 0

    for i in range(n):
        sum_x += x[i]
        sum_y += y[i]
        sum_xy += x[i] * y[i]
        sum_x2 += x[i] ** 2

    x_mean = sum_x / n
    y_mean = sum_y / n

    a = ((y_mean * sum_x2) - (x_mean * sum_xy)) / (sum_x2 - (n * (x_mean ** 2)))
    b = (sum_xy - (x_mean * sum_y)) / (sum_x2 - (n * (x_mean ** 2)))
                                  
    print("The straight line equation: y = (%f) + (%f)x\n" % (a, b))

straight_line_equation(x, y)

bestfit_y = []

for i in range(len(x)):
    bestfit_y.append((1.066667) + (0.500000) * x[i])

plt.title("x / y")
plt.xlabel("x", loc = "right")
plt.ylabel("y", loc = "top")
plt.grid()
plt.scatter(x, y, color = 'k', alpha = 0.5)
plt.plot(x, bestfit_y, color = "green")
plt.plot(x, y)
plt.legend(['scatter', 'bestfit', 'allfit'])
plt.show()
```
