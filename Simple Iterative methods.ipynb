```python
"""
Suppose 2x^2 - 5x + 3 = 0, x0 = 0

Rearraning the equation to give the x subject:

x = (2 * (x ** 2) + 3) / 5          or         x = np.sqrt(((5 * x) - 3) / 2)  

x1 = (2 * (x0 ** 2) + 3) / 5 = (2 * (0 ** 2) + 3) / 5 = 0.6

x2 = (2 * (x1 ** 2) + 3) / 5 = (2 * (0.6 ** 2) + 3) / 5 = 0.744

x3 = (2 * (x2 ** 2) + 3) / 5 = (2 * (0.744 ** 2) + 3) / 5 = 0.8214144

and so on 

"""

x = 0 #initial value of x

print("No degree of accuracy unspecified:\n")
for i in range(100):
    x_new = (2 * (x ** 2) + 3) / 5
    print(f"x{i + 1}: ", x_new, "\n")
    if x_new == x:
        break
    x = x_new
    
```

```python
x = 0

print("Degree of accuracy specified to 6-decimal places:")
for i in range(100):
    x_new = (2 * (x ** 2) + 3) / 5
    print(f"x{i + 1}: ", x_new, "\n")
    if (x_new - x) < 0.000001:
        break
    x = x_new


"""
The Degree of accuracy

_To satisfy the condition of exact equality between numerical values is extremely difficult if not impossible.

_So, the normal practice in numerical methods is to specify the degree of accuracy required in the solution.
 It reperesents the acceptable absolute maximum difference between values of the variable of the equation resulted
 from the two successive iterations, in other words, the difference between the old and the new values.
"""
```
