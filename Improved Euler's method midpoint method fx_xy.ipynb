```python
import pandas as pd
import matplotlib.pyplot as plt
import math

def midpoint_euler(f, t0, y0, t_end, h):
    t = t0
    y = y0
    results = []
    while t <= t_end:
        k1 = h * f(t, y)
        k2 = h * f(t + (h/2), y + (k1/2))
        results.append([t, y, f(t, y), k2])
        y = y + k2
        t = t + h
    df = pd.DataFrame(results, columns=['t', 'y', 'dy/dt', 'hdy/dt'])
    print(df)
    
    plt.plot(df['t'], df['y'])
    plt.plot(df['t'], df['dy/dt'])
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend(['function', 'derivative'])
    plt.title('Euler Method')
    plt.grid()
    plt.show()
    return y

# Test case
dy_dt = lambda t, y : (t ** 2) - 3 * t + y

t0 = 0
y0 = 1
t_end = 0.02
h = 0.01

solution = midpoint_euler(dy_dt, t0, y0, t_end, h)
print("Solution at t =", t_end, ":", solution)
```
